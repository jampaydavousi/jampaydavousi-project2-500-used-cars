```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Import the CSV file from the first data source and extract it into DataFrame
vehicle_file = "cleaned_data_500_2014.csv"
vehicle_df = pd.read_csv(vehicle_file)
vehicle_df
```

```python
#Add a unique id column
#vehicle_df = vehicle_df.reset_index(drop=True)
#vehicle_df
```

```python
#rename the index column to id as it would map to the id column of its respective SQL table 
vehicle_df2 = vehicle_df.rename(columns={'index': 'id'})
vehicle_df2
```

```python
vehicle_df2 = vehicle_df.rename(columns={'index': 'id',
                                         'Price': 'price', 
                                         'City':'city',
                                         'State': 'state',
                                         'Mileage': 'mileage',
                                         'Vin' : 'vin',
                                         'MSRP' : 'msrp',
                                         'Depreciation':'depreciation',
                                         'Latitude': 'latitude',
                                         'Longitude': 'longitude'
})
vehicle_df2
```

```python
###############################################################################################################
#Go to PostgreSQL and create the database vehicle_db, create a table corresponding to the data frame, 
#and create the column names for table exactly as they were renamed for the SQL convention
#in the above data frame.
###############################################################################################################
```

```python
#create a connection pandas and to the SQL environment - in this case PostgreSQL
rds_connection_string = "postgres:NIM2006@localhost:5432/vehicle_db_500_2014"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Once the table are created on the PostgreSQL side, verify the connection is working by recalling 
#the table name that is created in PostGreSQL inside the vehicle_db database.
engine.table_names()
```

```python
#Load the information from the dataframe to its respective table in postgreSQL
vehicle_df2.to_sql(name='vehicle_500_2014', con=engine, if_exists='append', index=False)
```
