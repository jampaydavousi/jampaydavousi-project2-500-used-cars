```python
#import csv with the VIN listings
vehicle_file = "filtered_cars_2014_500.csv"
vehicle_df = pd.read_csv(vehicle_file)
vehicle_df
```

```python
url2 = "https://vpic.nhtsa.dot.gov/api/vehicles/decodevin/"
```

```python
# loop through VIN’s to and call the code for API queries to get the results and append it to 3 lists.
for i in vehicle_df['Vin']:
    query = f"{url2}{i}?format=json"
    response = requests.get(query).json()
    
    make.append(response['Results'][6]['Value'])
    model.append(response['Results'][8]['Value'])
    year.append(response['Results'][9]['Value'])
                
```

```python
#https://www.searchquarry.com/vin-decoder/results.php?trackstat=id-21078-&sqtb=VIN&co=ft&vin=WBAUP7C56CVP22994
url4 = 'https://www.searchquarry.com/vin-decoder/results.php?trackstat=id-21078-&sqtb=VIN&co=ft&vin='

#browser.visit(url4)
#html4 = browser.html
#soup = BeautifulSoup(html4, 'html.parser')
```

```python
for j in vehicle_df['Vin']:
    query2 = f"{url4}{j}"
    browser.visit(query2)
    html4 = browser.html
    soup = BeautifulSoup(html4, 'html.parser')
    msrps = soup.find_all('td', class_='table_value')
    clean_msrp = msrps[66].text
    #print(clean_msrp)
    stripped_msrp = clean_msrp.strip()
    #print(clean_msrp.strip())
    msrp.append(stripped_msrp)
```

```python
# select required columns from the csv file
new_vehicle_df = vehicle_df[['Price','City','State','Mileage','Vin']]
new_vehicle_df.head()
```

```python
new_vehicle_df['MSRP'] = msrp
new_vehicle_df.head()
```

```python
new_vehicle_df['MSRP'] = new_vehicle_df['MSRP'].str.replace('$','')
new_vehicle_df
```

```python
new_vehicle_df['MSRP'] = new_vehicle_df['MSRP'].str.replace(',','')
new_vehicle_df
```

```python
new_vehicle_df2 = new_vehicle_df 
```

```python
new_vehicle_df2.head()
```

```python
new_vehicle_df['MSRP'] = pd.to_numeric(new_vehicle_df['MSRP'], errors ='coerce').fillna(0).astype(np.int64)
new_vehicle_df.head()
```

```python
new_vehicle_df['MSRP'] = new_vehicle_df['MSRP'].astype(int)
new_vehicle_df
```

```python
new_vehicle_df['Price'] = new_vehicle_df['Price'].astype(int)
new_vehicle_df
```

```python
# append retrieved data
new_vehicle_df['make'] = make
new_vehicle_df['model'] = model
new_vehicle_df['year'] = year
```

```python
# check new dataframe
new_vehicle_df
```

```python
#new_vehicle_df['MSRP'].describe
new_vehicle_df['Depreciation'] = (new_vehicle_df['MSRP'] - new_vehicle_df['Price'])*100/new_vehicle_df['MSRP']
new_vehicle_df
```

```python
new_vehicle_df['Depreciation'] = new_vehicle_df['Depreciation'].round(decimals=2)
new_vehicle_df
```

```python
url5 = "http://api.openweathermap.org/data/2.5/weather?"
```

```python
cities = new_vehicle_df['City']
lat = []
lon = []
country = []
for index, row in new_vehicle_df.iterrows():
   city = row["City"]
   query_url = f"{url5}appid=34154490158887b53402ad370d4af371&q={city},US"
   response = requests.get(query_url).json()
   try:
       country.append(response['sys']['country'])
       lat.append(response['coord']['lat'])
       lon.append(response['coord']['lon'])
   except KeyError:
       country.append(np.nan)
       lat.append(np.nan)
       lon.append(np.nan)
```

```python
new_vehicle_df["Latitude"] = lat
new_vehicle_df["Longitude"] = lon
```

```python
#Add a unique id column
new_vehicle_df = new_vehicle_df.reset_index(drop=False)
new_vehicle_df
```
